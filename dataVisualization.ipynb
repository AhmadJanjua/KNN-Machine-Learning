```python
import numpy as np 
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read in the data from the excel file.
df = pd.read_excel("data.xlsx")
df
```

```python
# Create the classification based on male and female gender
df = df.sort_values('Gender', ignore_index=True)
freq_m , freq_f = df.Gender.value_counts()
freq_f, freq_m
```

```python
# Data visualization using subplots
fig, axs = plt.subplots(figsize = (19,15))

axs1 = plt.subplot2grid( shape= (3,3), loc = (0,0))
axs2 = plt.subplot2grid( shape= (3,3), loc = (0,1))
axs3 = plt.subplot2grid( shape= (3,3), loc = (0,2))
axs4 = plt.subplot2grid( shape= (3,3), loc = (1,0))
axs5 = plt.subplot2grid( shape= (3,3), loc = (1,1))
axs6 = plt.subplot2grid( shape= (3,3), loc = (1,2))
axs7 = plt.subplot2grid( shape= (3,3), loc = (2,0))
axs8 = plt.subplot2grid( shape= (3,3), loc = (2,1))
axs9 = plt.subplot2grid( shape= (3,3), loc = (2,2))
plt.tight_layout()

#Histogram 1
axs1.hist( df.iloc[ 0:freq_f, 0], edgecolor = 'brown', fc = 'none', label = 'Female')
axs1.hist( df.iloc[ freq_f:freq_f+freq_m, 0], edgecolor = 'green', fc = 'none', label = 'Male')

axs1.set_xlabel('Height (cm)    (a)')
axs1.set_ylabel('Frequency')
axs1.legend()

#Scatterplot 1
axs2.scatter( df.iloc[0:freq_f,1], df.iloc[0:freq_f,0], edgecolor = 'brown', fc = 'none', label = 'Female')
axs2.scatter( df.iloc[freq_f:freq_f+freq_m,1], df.iloc[freq_f:freq_f+freq_m,0], edgecolor = 'green', fc = 'none', label = 'Male')

axs2.set_xlabel('Shoe Size US   (b)')
axs2.set_ylabel('Height (cm)')
axs2.legend()

#Scatterplot 2
axs3.scatter( df.iloc[0:freq_f,2], df.iloc[0:freq_f,0], edgecolor = 'brown', fc = 'none', label = 'Female')
axs3.scatter( df.iloc[freq_f:freq_f+freq_m,2], df.iloc[freq_f:freq_f+freq_m,0], edgecolor = 'green', fc = 'none', label = 'Male')

axs3.set_xlabel('Pets   (c)')
axs3.set_ylabel('Height (cm)')
axs3.legend()

#Scatterplot 3
axs4.scatter( df.iloc[0:freq_f,0], df.iloc[0:freq_f,1], edgecolor = 'brown', fc = 'none', label = 'Female')
axs4.scatter( df.iloc[freq_f:freq_f+freq_m,0], df.iloc[freq_f:freq_f+freq_m,1], edgecolor = 'green', fc = 'none', label = 'Male')

axs4.set_xlabel('Height (cm)    (d)')
axs4.set_ylabel('Shoe Size US')
axs4.legend()

#Histogram 2
axs5.hist( df.iloc[ 0:freq_f, 1], edgecolor = 'brown', fc = 'none', label = 'Female')
axs5.hist( df.iloc[ freq_f:freq_f+freq_m, 1], edgecolor = 'green', fc = 'none', label = 'Male')

axs5.set_xlabel('Shoe Size US   (e)')
axs5.set_ylabel('Frequency')
axs5.legend()

#Scatterplot 4
axs6.scatter( df.iloc[0:freq_f,2], df.iloc[0:freq_f,1], edgecolor = 'brown', fc = 'none', label = 'Female')
axs6.scatter( df.iloc[freq_f:freq_f+freq_m,2], df.iloc[freq_f:freq_f+freq_m,1], edgecolor = 'green', fc = 'none', label = 'Male')

axs6.set_xlabel('Pets   (f)')
axs6.set_ylabel('Shoe Size US')
axs6.legend()

#Scatterplot 5
axs7.scatter( df.iloc[0:freq_f,0], df.iloc[0:freq_f,2], edgecolor = 'brown', fc = 'none', label = 'Female')
axs7.scatter( df.iloc[freq_f:freq_f+freq_m,0], df.iloc[freq_f:freq_f+freq_m,2], edgecolor = 'green', fc = 'none', label = 'Male')

axs7.set_xlabel('Height (cm)    (g)')
axs7.set_ylabel('Pets')
axs7.legend()

#Scatterplot 6
axs8.scatter( df.iloc[0:freq_f,1], df.iloc[0:freq_f,2], edgecolor = 'brown', fc = 'none', label = 'Female')
axs8.scatter( df.iloc[freq_f:freq_f+freq_m,1], df.iloc[freq_f:freq_f+freq_m,2], edgecolor = 'green', fc = 'none', label = 'Male')
axs8.set_xlabel('Shoe Size US   (h)')
axs8.set_ylabel('Pets')
axs8.legend()

#Histogram 3
axs9.hist( df.iloc[ 0:freq_f, 2], edgecolor = 'brown', fc = 'none', label = 'Female')
axs9.hist( df.iloc[ freq_f:freq_f+freq_m, 2], edgecolor = 'green', fc = 'none', label = 'Male')

axs9.set_xlabel('Pets   (i)')
axs9.set_ylabel('Frequency')
axs9.legend()
plt.show()
```
